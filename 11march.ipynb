```python
# Q2: Differentiate between one-tailed and two-tailed tests.
# Ans 2:In hypothesis testing, a one-tailed test (also known as a directional test) is a statistical test in which the critical region
# is on only one side of the sampling distribution. This means that the hypothesis is tested in only one direction, either positive 
# or negative. A one-tailed test is used when there is a specific directional hypothesis, and the researcher is interested in testing
# whether the sample mean is greater than or less than a particular value.

# On the other hand, a two-tailed test (also known as a non-directional test) is a statistical test in which the critical region is on
# both sides of the sampling distribution. This means that the hypothesis is tested in both directions, positive and negative. A two-tailed 
# test is used when there is no specific directional hypothesis, and the researcher is interested in testing whether the sample mean is 
# simply different from a particular value
```

```python
# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
# each type of error.
# Ans3:In hypothesis testing, Type 1 and Type 2 errors are two types of errors that can occur while making decisions 
# about the null hypothesis.

# Type 1 error occurs when the null hypothesis is rejected even when it is true. This means that the researcher concludes 
# that there is a significant effect or difference when there is not one. Type 1 errors are also called false positives. 
# The probability of making a Type 1 error is denoted by the Greek letter alpha (α) and is usually set at 0.05 or 0.01.

# For example, in a clinical trial of a new drug, a Type 1 error would occur if the researcher concludes that the drug is 
# effective in treating a disease when it is not actually effective. This error would result in the drug being approved for
# use when it is not actually beneficial.

# Type 2 error, on the other hand, occurs when the null hypothesis is not rejected when it is false. This means that the
# researcher fails to detect a significant effect or difference when one exists. Type 2 errors are also called false negatives.
# The probability of making a Type 2 error is denoted by the Greek letter beta (β).

# For example, in a clinical trial of a new drug, a Type 2 error would occur if the researcher concludes that the drug is
# not effective in treating a disease when it is actually effective. This error would result in the drug not being approved
# for use when it could have been beneficial.

# It is important to note that the probability of making a Type 1 error and a Type 2 error are inversely related. That is, 
# as the probability of making one type of error decreases, the probability of making the other type of error increases.
# Therefore, researchers need to balance the risks of making these errors when designing their hypothesis tests.
```

```python
# Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
# Ans 5:A confidence interval is a range of values that provides an estimate of the true population parameter with a certain level of 
# confidence. In statistical inference, it is used to estimate an unknown population parameter based on the sample statistic. It 
# represents a range of values within which the true population parameter is likely to fall.

# The formula to calculate a confidence interval is:

# CI = X ± Z * (σ / √n)

# Where CI is the confidence interval, X is the sample mean, Z is the critical value obtained from the standard normal distribution
# table for the desired level of confidence, σ is the population standard deviation (if known), and n is the sample size.

# For example, suppose we want to estimate the average height of students in a school with 95% confidence based on a sample of 50
# students. If the sample mean height is 68 inches and the population standard deviation is 3 inches, we can calculate the confidence 
# interval as follows:

# CI = 68 ± 1.96 * (3 / √50)
# CI = 68 ± 0.84
# CI = (67.16, 68.84)

# Therefore, we can say with 95% confidence that the true population mean height of students in the school falls between 67.16 and 
# 68.84 inches.
```

```python
# Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
# event's probability and new evidence. Provide a sample problem and solution.
# Ans-
#  Problem:
 
#  Suppose a company produces two different types of electronic components, A and B. Component A has a defect rate of 5%, while component B has a defect rate of 10%. A customer reports receiving a defective component, but it is not known whether it was type A or type B. What is the probability that the defective component was type A?

# Solution:
    
# Let D be the event that the component is defective, and let A and B be the events that the component is type A or type B, respectively. We want to find P(A|D), the probability that the component is type A given that it is defective.

# By Bayes' Theorem, we have:

#     P(A|D) = P(D|A) * P(A) / P(D)

# where P(D|A) is the probability that the component is defective given that it is type A, P(A) is the prior probability that the component is type A, and P(D) is the marginal probability of the component being defective.

# We know that P(D|A) = 0.05, since type A components have a 5% defect rate. We also know that P(A) = 0.5, since there are two types of components and they are equally likely to be selected. Finally, we can use the law of total probability to find P(D):

#     P(D) = P(D|A) * P(A) + P(D|B) * P(B)

# where P(D|B) is the probability that the component is defective given that it is type B, and P(B) is the prior probability that the component is type B. We know that P(D|B) = 0.1, and P(B) = 0.5, so we can substitute these values:

#     P(D) = 0.05 * 0.5 + 0.1 * 0.5 = 0.075

# Now we can plug all these values into Bayes' Theorem:

#     P(A|D) = 0.05 * 0.5 / 0.075 = 0.333

# Therefore, the probability that the defective component was type A is 0.333, or approximately 33.3%.
```

```python
# Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
# of 5. Interpret the results.
# Ans-
# To calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5, we need to use the following formula:
#95% confidence interval = sample mean ± 1.96 x standard error
# where the standard error is calculated as:
#     standard error = standard deviation / square root of sample size
# Assuming a sample size of 100, we can calculate the standard error as:
#     standard error = 5 / sqrt(100) = 0.5
# Substituting the values in the formula, we get:
#     95% confidence interval = 50 ± 1.96 x 0.5 = (49.01, 50.99)
# Interpretation: We can say with 95% confidence that the true population mean lies between 49.01 and 50.99.
```

```python
# Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
# Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
# Ans-
#     The margin of error is the maximum expected difference between the true population parameter and the estimated sample statistic. It is a measure of the precision of the estimate and is typically expressed as a percentage. The margin of error is calculated by multiplying the critical value of the confidence interval by the standard error of the sample mean.

# Sample size has a direct effect on the margin of error. As the sample size increases, the margin of error decreases. This is because a larger sample size results in a more representative sample of the population, which in turn reduces the variability of the sample mean and increases the precision of the estimate.

# For example, suppose a polling company wants to estimate the percentage of voters in a city who support a particular candidate. They take a random sample of 500 voters and find that 55% of them support the candidate. They want to calculate a 95% confidence interval for the true proportion of supporters in the population.

# Using the formula for the margin of error, we get:

# Margin of error = Z * (standard deviation / sqrt(sample size))
# Z = 1.96 (for a 95% confidence level)
# Standard deviation = sqrt(p * (1 - p) / n), where p is the sample proportion and n is the sample size
# p = 0.55
# n = 500

# Plugging in the values, we get:

#     Margin of error = 1.96 * (sqrt(0.55 * 0.45 / 500)) = 0.047

# Therefore, the 95% confidence interval for the true proportion of supporters is 0.55 +/- 0.047, or 0.503 to 0.597. This means we can be 95% confident that the true proportion of supporters in the population falls within this range.

# If the polling company were to increase their sample size to 1000, the margin of error would decrease, resulting in a smaller range for the confidence interval and a more precise estimate of the true proportion of supporters in the population.
```

```python
# Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
# of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
# significantly effective at a 95% confidence level using a t-test.
# Ans-
#     To determine if the new weight loss drug is significantly effective, we can use a one-sample t-test. The null hypothesis is that the true population mean weight loss is equal to zero, while the alternative hypothesis is that the true population mean weight loss is greater than zero. We will use a 95% confidence level, which corresponds to a significance level of 0.05.

# The test statistic for a one-sample t-test is given by:

#      t = (x̄ - μ) / (s / sqrt(n))

# where x̄ is the sample mean, μ is the population mean (in this case, assumed to be zero), s is the sample standard deviation, and n is the sample size.

# Plugging in the values from the problem, we get:

#      t = (6 - 0) / (2.5 / sqrt(50))
#     t = 13.42

# The degrees of freedom for this test are (n - 1) = 49.

# Using a t-distribution table or a calculator, we find the critical t-value for a one-tailed test with 49 degrees of freedom at a significance level of 0.05 to be 1.676.

# Since our calculated t-value (13.42) is greater than the critical t-value (1.676), we can reject the null hypothesis and conclude that the new weight loss drug is significantly effective at a 95% confidence level.
```

```python
# Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
# confidence interval for the true proportion of people who are satisfied with their job.
# Ans-
#    To calculate the confidence interval for a proportion, we can use the formula:

# CI = p ± zsqrt((p(1-p))/n)

# where:

# CI: confidence interval
# p: sample proportion
# z: z-score corresponding to the desired confidence level (in this case, 1.96 for 95% confidence)
# n: sample size
# Using the given values, we can calculate the confidence interval as follows:

# p = 0.65
# n = 500
# z = 1.96

# CI = 0.65 ± 1.96sqrt((0.65(1-0.65))/500)

# CI = 0.65 ± 0.049

# CI = (0.601, 0.699)

# Therefore, we can be 95% confident that the true proportion of people who are satisfied with their job is between 0.601 and 0.699.
```

```python
# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
# Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
# with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
# significant difference in student performance using a t-test with a significance level of 0.01.

# Ans-
#     To conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance, we can use a two-sample t-test. The null hypothesis is that there is no significant difference between the two teaching methods, while the alternative hypothesis is that there is a significant difference.

#        H0: μA = μB
#         Ha: μA ≠ μB

# Where μA is the population mean score for sample A and μB is the population mean score for sample B.

# Assuming equal variances, we can calculate the pooled standard deviation as follows:

# Sp = sqrt(((nA-1)*sA^2 + (nB-1)*sB^2) / (nA + nB - 2))
# where nA = nB = sample sizes of A and B
# sA = standard deviation of sample A
# sB = standard deviation of sample B

# Plugging in the values, we get:

# Sp = sqrt(((20-1)*6^2 + (20-1)*5^2) / (20+20-2)) = 5.527

# Now, we can calculate the t-statistic using the following formula:

# t = (x̄A - x̄B) / (Sp * sqrt(1/nA + 1/nB))
# where x̄A = sample mean score for sample A
# x̄B = sample mean score for sample B

# Plugging in the values, we get:

# t = (85 - 82) / (5.527 * sqrt(1/20 + 1/20)) = 3.25

# Using a t-table with degrees of freedom of 38 (nA + nB - 2), we find that the critical t-value for a two-tailed test with a significance level of 0.01 is approximately 2.704.

# Since our calculated t-value of 3.25 is greater than the critical t-value of 2.704, we reject the null hypothesis and conclude that there is a significant difference between the two teaching methods in terms of student performance.

# In other words, we can say that there is strong evidence to suggest that one of the teaching methods is better than the other in terms of student performance.
```

```python
# Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average
# reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
# determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

# Ans-
#     To conduct a hypothesis test, we need to state our null and alternative hypotheses:

# Null hypothesis: The caffeine has no significant effect on reaction time. The population mean reaction time is equal to or greater than 0.25 seconds.
# Alternative hypothesis: The caffeine has a significant effect on reaction time. The population mean reaction time is less than 0.25 seconds.
# We will use a one-tailed t-test with a significance level of 0.10 (90% confidence level) and a sample size of 30. The degrees of freedom are 29 (sample size minus 1).

# We can calculate the t-statistic as:

# t = (sample mean - hypothesized population mean) / (standard deviation / sqrt(sample size))

# t = (0.25 - 0) / (0.05 / sqrt(30)) = 6.708

# Using a t-distribution table or calculator with 29 degrees of freedom and a one-tailed test with a significance level of 0.10, we find the critical value to be -1.311.

# Since the calculated t-statistic (6.708) is greater than the critical value (-1.311), we reject the null hypothesis and conclude that the caffeine has a significant effect on reaction time at a 90% confidence level.

# Therefore, we can say that there is strong evidence to suggest that caffeine has a positive effect on reaction time, making individuals more alert and responsive.
```
